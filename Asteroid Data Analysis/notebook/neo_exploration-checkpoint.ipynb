```python
import pandas as pd
df = pd.read_csv("data/neos.csv")
print("Rows:", len(df))
print("Unique asteroids:", df["id"].nunique())
print("Hazardous:", df["hazardous"].sum())
print("Columns:", len(df.columns))
print("Date range:", df["date"].min(), "->", df["date"].max())
```

```python
df.groupby("id").size().sort_values(ascending=False).head()
```
